# README

This takes the data generated from gen_2DWindow_data.py and converts it into graphs. This experiment is taking 12 curves where there are windows around the null points. These windows around the null points are varied to show how close we need to be to the null point to have improved MSE beyond the conventional method. This code is changed from the npWindow experiment in that both windows vary in the central location. We also allow the SNR to vary.

# Functions

```python
def to_readable_2DWindow(file):
    df = pd.read_pickle(file)
    ALL_CENTER1 = np.stack(df["TI1*g"].values) #shape: 
    ALL_CENTER2 = np.stack(df["TI2*g"].values) #shape: 
    ALL_TI = np.stack(df["TI_DATA"].values) #shape:  #shape: 
    ALL_MSE = np.stack(df["MSE"].values) #shape: 
    ALL_VAR = np.stack(df["var"].values) #shape: 
    ALL_BIAS = np.stack(df["bias"].values) #shape: 

    ALL_pEst_AIC = np.stack(df["pEst_AIC"].values) #shape: 
    ALL_pEst_cf = np.stack(df["pEst_cf"].values) #shape: 
    ALL_pEst_cvn = np.stack(df['pEst_cvn'].values)
    
    return ALL_CENTER1, ALL_CENTER2, ALL_TI, ALL_MSE, ALL_VAR, ALL_BIAS, ALL_pEst_AIC, ALL_pEst_cf, ALL_pEst_cvn

def to_readable_SR(file,p_oi):
    df = pd.read_pickle(file)
    ALL_MSE = np.stack(df["MSE"].values)[0,:,p_oi] #shape: 
    ALL_VAR = np.stack(df["var"].values)[0,:,p_oi] #shape: 
    ALL_BIAS = np.stack(df["bias"].values)[0,:,p_oi] #shape: 

    ALL_pEst_AIC = np.stack(df["pEst_AIC"].values)[0,:] #shape: 
    ALL_pEst_cf = np.stack(df["pEst_cf"].values)[0,:] #shape: 
    ALL_pEst_cvn = np.stack(df["pEst_cvn"].values)[0,:] #shape: 
    
    return ALL_MSE, ALL_VAR, ALL_BIAS, ALL_pEst_AIC, ALL_pEst_cf, ALL_pEst_cvn

def int_lengh_AIC(R0, diff_TE, SNR):
    return (2*diff_TE/(R0*SNR**2))**(1/2)

def calc_MSE(paramStore, true_params, clipped = False):
    varMat = np.var(paramStore, axis=0)
    biMat = np.mean(paramStore, axis = 0) - true_params  #E[p_hat] - p_true
    MSEMat = varMat + biMat**2
    if clipped:
        return MSEMat[-4:], biMat[-4:], varMat[-4:]
    return MSEMat, varMat, biMat

def _get_colors(num_colors):
    colors = []
    for i in np.arange(0., 360., 360. / num_colors):
        hue = i/360.
        lightness = (50 + np.random.rand() *10)/100.
        saturation = (90 + np.random.rand() * 10)/100.
        colors.append(colorsys.hls_to_rgb(hue, lightness, saturation))
    return colors
```

```python
def trapz_3d(x, y, z):
    area = 0
    for iY in range(z.shape[0]-1):
        for iX in range(z.shape[1]-1):
            four_corners = [z[iY,iX], z[iY+1,iX], z[iY,iX+1], z[iY+1,iX+1]]
            corner_sig = np.mean(four_corners)
            rect_base = (x[iX+1]-x[iX])*(y[iY+1]-y[iY])
            area = area + corner_sig*rect_base
    return area
```

```python
def prepare_data(p_oi):
    # ordering = ["ESPRESSO", "Conventional", "Curve Fit"]
    MSE_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices),3))
    var_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices),3))
    bias_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices),3))

    TI_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices), 12))

    pEst_AIC_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices), num_NR, len(ParamTitle_6p)))
    pEst_cf_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices), num_NR, len(ParamTitle_6p)))
    pEst_cvn_ordered = np.zeros((len(TI1g_indices), len(TI2g_indices), num_NR, len(ParamTitle_6p)))

    data_name = f"{cwd_full}{base_file}/{exp_file}/{data_file}/reassignExp_run{run_number}_{exp_date}.pkl"

    print("Retrieving data for " + data_name)
    NP1, NP2, TI_DATA,  MSE, VAR, BIAS, pEst_AIC, pEst_cf, pEst_cvn = to_readable_2DWindow(data_name)

    for index in range(len(MSE)):

        temp_np1 = NP1[index]
        temp_np2 = NP2[index]

        TI1_int_index = np.where(temp_np1==TI_FULL)[0][0]
        TI2_int_index = np.where(temp_np2==TI_FULL)[0][0]

        np1_index = np.where(TI1_int_index == TI1g_indices)[0][0]
        np2_index = np.where(TI2_int_index == TI2g_indices)[0][0]

        MSE_ordered[np1_index, np2_index, :] = MSE[index,:,p_oi]
        var_ordered[np1_index, np2_index, :] = VAR[index,:,p_oi]
        bias_ordered[np1_index, np2_index, :] = BIAS[index,:,p_oi]

        TI_ordered[np1_index, np2_index, :] = TI_DATA[index,:]
        pEst_AIC_ordered[np1_index, np2_index, :, :] = pEst_AIC[index,:,:]
        pEst_cf_ordered[np1_index, np2_index, :, :] = pEst_cf[index,:,:]
        pEst_cvn_ordered[np1_index, np2_index, :, :] = pEst_cvn[index,:,:]

    return MSE_ordered, var_ordered, bias_ordered, TI_ordered, pEst_AIC_ordered, pEst_cf_ordered, pEst_cvn_ordered
```

```python
sp_map = plt.get_cmap('seismic_r')
sp_map.set_extremes(under = 'orange')
extend_opt = "neither"      #max or neither

def gen_patchwork_panel(fig, ax, imp_mat, ex_title='', cb_range = 0):
    ######## bottom left

    if cb_range == 0:
        pos_max = np.max(imp_mat)
    else:
        pos_max = cb_range

    plt.rc('font', size = 12)
    ax.set_title(f"{ex_title}[{T11:.0f},{T12:.0f},{c1:.2f},{c2:.2f},{T21:.0f},{T22:.0f}]")
    imc1 = ax.imshow(imp_mat[:,:], cmap = sp_map, vmin = -1*pos_max, vmax = pos_max)

    # ax.set_title('Improvement in ' + param_name_list[POI] + ' MSE')
    ax.set_yticks(np.arange(len(TI1g_indices)))
    ax.set_xticks(np.arange(len(TI2g_indices)))

    temp_x_labels = ["" for a in TI2g_indices]
    x_array = [0,len(TI2g_indices)//4,len(TI2g_indices)//2, 3*len(TI2g_indices)//4, -1]
    for elem in x_array:
        temp_x_labels[elem] = f"{TI_FULL[TI2g_indices[elem]]:0.1f}"
    ax.set_xticklabels(temp_x_labels)

    temp_y_labels = ["" for a in TI1g_indices]
    y_array = [0,len(TI1g_indices)//4,len(TI1g_indices)//2, 3*len(TI1g_indices)//4, -1]
    for elem in y_array:
        temp_y_labels[elem] = f"{TI_FULL[TI1g_indices[elem]]:0.1f}"
    ax.set_yticklabels(temp_y_labels)

    ax.set_ylabel(r'TI$_1^*$ Guess Offset')
    ax.set_xlabel(r'TI$_2^*$ Guess Offset')
    plt.rc('font', size = 12)
    cbar = fig.colorbar(imc1, ax=ax, fraction=0.046, pad=0.2, location = 'bottom', extend = extend_opt)#, format = '%.2e')
    cbar.formatter.set_powerlimits((0,0))

    return
```
